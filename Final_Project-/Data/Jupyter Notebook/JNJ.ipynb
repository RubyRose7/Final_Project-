```python
import pandas as pd
import numpy as np
```

```python
JNJ = 'JNJ.csv'
```

```python
jnj_df = pd.read_csv(JNJ)
jnj_df
```

```python
jnj1_df=jnj_df.fillna(0)
jnj1_df
```

```python
# Check any missing data
for col in jnj1_df.columns:
    missing_data = np.mean(jnj1_df[col].isnull())
    print('{} - {}%'.format(col, missing_data))
```

```python
# Select the columns to keep

new_jnj1_df = jnj1_df[["Date", "Open", "Close", "Volume"]]
new_jnj1_df
```
