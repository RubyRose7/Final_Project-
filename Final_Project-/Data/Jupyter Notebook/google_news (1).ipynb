```python
def get_titles(search):
    stories = []
    gn = GoogleNews(country = 'US')
    search = gn.search(search, when='8d')
    newsitem = search['entries']
    for item in newsitem:
        story = {
            'title': item.title,
            'link': item.link
        }
        stories.append(story)
    return stories 

get_titles('JNJ')
```

```python
def get_titles(search):
    stories = []
    stories_titles=[]
    stories_links=[]
    gn = GoogleNews(country = 'US')
    search = gn.search(search, when='8d')
    newsitem = search['entries']
    for item in newsitem:
        stories_titles.append(item.title)
        stories_links.append(item.link)
    story = {
        'Titles': stories_titles,
        'Links': stories_links
    }
#         stories.append(story)
    return story


#df= get_titles('Amazon')
jnj_df = pd.DataFrame(get_titles('JNJ'))
jnj_df
```

```python
# Push the remade DataFrame to a new CSV file
jnj_df.to_csv("FinalJNJ.csv")
```
