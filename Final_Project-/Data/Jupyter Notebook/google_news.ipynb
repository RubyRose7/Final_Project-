### Sort by News Title and Link

```python
def get_titles(search):
    stories = []
    gn = GoogleNews(country = 'US')
    search = gn.search(search, when= '')
    newsitem = search['entries']
    for item in newsitem:
        story = {
            'title': item.title,
            'link': item.link
        }
        stories.append(story)
    return stories 

get_titles('JNJ')
```
